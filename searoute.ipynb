```python
from searoute import searoute
import json
from IPython.display import IFrame, display
import os

# Cell 2: Functions

def get_coordinates():
    print("Enter coordinates as decimal degrees.")
    start_lat = float(input("Start Latitude: "))
    start_lon = float(input("Start Longitude: "))
    end_lat = float(input("End Latitude: "))
    end_lon = float(input("End Longitude: "))
    return (start_lat, start_lon), (end_lat, end_lon)

def generate_leaflet_html(route_coords, start, end):
    route_json = json.dumps(route_coords)

    html = f"""
<!DOCTYPE html>
<html>
<head>
  <title>SeaRoute Map</title>
  <meta charset="utf-8" />
  <meta name="viewport" content="width=device-width, initial-scale=1.0">

  <link rel="stylesheet" href="https://unpkg.com/leaflet/dist/leaflet.css" />
  <script src="https://unpkg.com/leaflet/dist/leaflet.js"></script>

  <style>
    #map {{ height: 100vh; }}
  </style>
</head>
<body>
  <div id="map"></div>

  <script>
    var map = L.map('map').setView([{start[0]}, {start[1]}], 4);

    L.tileLayer('https://{{s}}.tile.openstreetmap.org/{{z}}/{{x}}/{{y}}.png', {{
      maxZoom: 18,
    }}).addTo(map);

    var route = {route_json};

    var latlngs = route.map(function(coord) {{
        return [coord[1], coord[0]];  // swap lon-lat to lat-lon for Leaflet
    }});

    L.polyline(latlngs, {{
      color: 'blue',
      weight: 3
    }}).addTo(map);

    L.marker([{start[0]}, {start[1]}]).addTo(map)
      .bindPopup('Start')
      .openPopup();

    L.marker([{end[0]}, {end[1]}]).addTo(map)
      .bindPopup('End');
  </script>
</body>
</html>
"""
    filename = "searoute_map.html"
    with open(filename, "w", encoding="utf-8") as f:
        f.write(html)

    print(f"✅ Leaflet map saved as '{filename}' in: {os.getcwd()}")
    return IFrame(src=filename, width="100%", height=600)
# Cell 3: Run the tool

start, end = get_coordinates()

# searoute expects (lon, lat)
route = searoute((start[1], start[0]), (end[1], end[0]))


if not route:
    print("❌ Route not found or invalid coordinates.")
else:
    display(generate_leaflet_html(route, start, end))
```

```python
import folium
from searoute import searoute

# Your precise port coordinates
port_klang = (101.39189318037145, 2.9984905997652835)        # (lon, lat)
singapore = (103.77255681534537, 1.284075259506207)          # (lon, lat)

# Calculate route using searoute (expects (lon, lat))
result = searoute(port_klang, singapore)

# Extract route coordinates: list of [lon, lat]
route_coords = result["geometry"]["coordinates"]

# Force exact start and end
route_coords.insert(0, list(port_klang))
route_coords.append(list(singapore))

# Convert to [lat, lon] for folium
latlon_route = [(lat, lon) for lon, lat in route_coords]

# Center map between both ports
center_lat = (port_klang[1] + singapore[1]) / 2
center_lon = (port_klang[0] + singapore[0]) / 2
m = folium.Map(location=[center_lat, center_lon], zoom_start=8)

# Add route polyline
folium.PolyLine(latlon_route, color="blue", weight=5, tooltip="Ship Route").add_to(m)

# Add port markers
folium.Marker([2.9984905997652835, 101.39189318037145], tooltip="Port Klang", icon=folium.Icon(color='green')).add_to(m)
folium.Marker([1.284075259506207, 103.77255681534537], tooltip="Singapore", icon=folium.Icon(color='red')).add_to(m)

# Save map to HTML
m
```

```python
import random
import math

# Fitness function: adjust this based on your use case
def fitness(x):
    return x * math.sin(10 * math.pi * x) + 1.0

# Generate a random chromosome (float between 0 and 1)
def generate_chromosome():
    return random.uniform(0, 1)

# Crossover: blend parent genes
def crossover(parent1, parent2):
    alpha = random.random()
    return alpha * parent1 + (1 - alpha) * parent2

# Mutation: add a small random value
def mutate(chromosome, mutation_rate=0.1):
    if random.random() < mutation_rate:
        mutation = random.uniform(-0.1, 0.1)
        chromosome = min(max(chromosome + mutation, 0), 1)
    return chromosome

# Genetic Algorithm
def genetic_algorithm(pop_size=20, generations=100):
    population = [generate_chromosome() for _ in range(pop_size)]

    for gen in range(generations):
        # Evaluate fitness
        population.sort(key=fitness, reverse=True)
        print(f"Gen {gen}: Best = {population[0]:.5f} | Fitness = {fitness(population[0]):.5f}")

        # Select top 50% as parents
        parents = population[:pop_size // 2]

        # Generate new population via crossover + mutation
        children = []
        while len(children) < pop_size:
            p1, p2 = random.sample(parents, 2)
            child = crossover(p1, p2)
            child = mutate(child)
            children.append(child)

        population = children

    # Final best solution
    best = max(population, key=fitness)
    print(f"\nBest solution: x = {best:.5f}, fitness = {fitness(best):.5f}")
    return best

# Run it
genetic_algorithm()
```
